```python
# Assuming your DataFrame is named df and 'state' is the column containing the stages
index_names = df.index.tolist()
# Create empty lists to store features and target variable
X_samples = []
y_samples = []
count=0
for id in index_names:
  id_to_find = id[:12]
  # Check if the ID is in the 'your_id_column' column
  if id_to_find in lb['bcr_patient_barcode'].values:
    # Get the stage for the specified ID
    stage = lb.loc[lb['bcr_patient_barcode'] == id_to_find, 'ajcc_pathologic_stage'].values[0]
    count=count+1
    print(f"{count}_Stage for {id_to_find}: {stage}")
    # Append the sample to the lists
    print(df.loc[id])
    X_samples.append(df.loc[id])
    y_samples.append(stage)
# Convert the lists to DataFrames
X_dataset = pd.DataFrame(X_samples)
y_dataset = pd.DataFrame(y_samples)
```

```python
# Assuming 'state' is the column containing the stages
# Create a mapping dictionary
stage_mapping = {
    'Stage I': 0,
    'Stage IA': 0,
    'Stage IB': 0,
    'Stage II': 0,
    'Stage IIA': 0,
    'Stage IIB': 0,
    'Stage III': 1,
    'Stage IIIA': 1,
    'Stage IIIB': 1,
    'Stage IIIC': 1,
    'Stage IV': 2,
    'Stage X': 3,
}

# Replace the stage names with integer values using the map function
y_dataset = y_dataset[0].replace(stage_mapping)
# Convert 'nan' values to a default integer (e.g., 0)
y_dataset = y_dataset.fillna(0).astype(int)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from collections import Counter  # To check class distribution

# Assuming X contains your features and y contains your labels
X = X_dataset  # Your feature data
y = y_dataset  # Your label data

# Display the class distribution before SMOTE
class_distribution_before = Counter(y)
print("Class Distribution Before SMOTE:", class_distribution_before)

# Determine the minority class size
minority_class_size = min(Counter(y).values())

# Ensure k_neighbors is at least 1
k_neighbors_value = max(1, minority_class_size - 1)

# Set n_neighbors to a value <= the number of samples in the minority class
smote = SMOTE(k_neighbors=k_neighbors_value, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Print the number of features before Lasso
print(f"Number of features before Lasso: {X_resampled.shape[1]}")

# Standardize the data before applying Lasso
scaler = StandardScaler()
X_resampled_scaled = scaler.fit_transform(X_resampled)

# Apply Lasso for feature selection
lasso = Lasso(alpha=0.01, random_state=42)  # You can adjust the alpha parameter as needed
lasso.fit(X_resampled_scaled, y_resampled)

# Get the indices of the features selected by Lasso (non-zero coefficients)
selected_features = np.where(lasso.coef_ != 0)[0]

# Filter the features in X_resampled based on the selected indices
X_resampled_lasso = X_resampled.iloc[:, selected_features]

# Now split the Lasso-selected features into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled_lasso, y_resampled, test_size=0.3, random_state=42)

# Display the class distribution after SMOTE and Lasso feature selection
class_distribution_after = Counter(y_train)
print("Class Distribution After SMOTE and Lasso:", class_distribution_after)

# Print the number of features selected by Lasso
print(f"Number of features selected by Lasso: {len(selected_features)}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from sklearn.metrics import precision_recall_curve, average_precision_score

# Create a RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the resampled data
model.fit(X_train, y_train)

# Make predictions on the original test set
predictions = model.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions,digits=6)
conf_matrix = confusion_matrix(y_test, predictions)

# Calculate specificity
specificity = conf_matrix[0, 0] / (conf_matrix[0, 0] + conf_matrix[0, 1])

# # Plot ROC curve
# Binarize the labels
y_test_binarized = label_binarize(y_test, classes=model.classes_)

# Fit the classifier
classifier = OneVsRestClassifier(model)
classifier.fit(X_train, y_train)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(model.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], classifier.predict_proba(X_test)[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves for each class
plt.figure(figsize=(10, 6))
for i in range(len(model.classes_)):
    plt.plot(fpr[i], tpr[i], lw=2, label='ROC curve of class {} (area = {:.2f})'.format(i, roc_auc[i]))

print(f"Accuracy: {accuracy}")
print(f"Specificity: {specificity}")
print(f"Classification Report:\n{report}")


plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()


# Compute precision-recall curve and average precision for each class
precision = dict()
recall = dict()
average_precision = dict()
for i in range(len(model.classes_)):
    precision[i], recall[i], _ = precision_recall_curve(y_test_binarized[:, i], classifier.predict_proba(X_test)[:, i])
    average_precision[i] = average_precision_score(y_test_binarized[:, i], classifier.predict_proba(X_test)[:, i])

# Plot precision-recall curves for each class
plt.figure(figsize=(10, 6))
for i in range(len(model.classes_)):
    plt.plot(recall[i], precision[i], lw=2, label='PR curve of class {} (AP = {:.2f})'.format(i, average_precision[i]))


plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()
```
